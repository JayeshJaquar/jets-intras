Imports and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load Data and fix Date parsing
df = pd.read_csv('combined_crypto_data.csv')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date']).sort_values(by=['coin', 'Date'])

print(f"Total rows: {len(df)} across {df['coin'].nunique()} unique coins.")
```

Feature Engineering


```python
def engineer_features(df):
    grouped = df.groupby('coin')

    df['return_1'] = grouped['Close'].pct_change(1)
    df['return_3'] = grouped['Close'].pct_change(3)
    df['volatility_7'] = grouped['return_1'].transform(lambda x: x.rolling(7).std())
    df['volume_change'] = grouped['Volume'].pct_change(1)

    df['ma_7'] = grouped['Close'].transform(lambda x: x.rolling(7).mean())
    df['ma_14'] = grouped['Close'].transform(lambda x: x.rolling(14).mean())
    df['ma_distance_7'] = (df['Close'] - df['ma_7']) / df['ma_7']
    df['ma_crossover'] = (df['ma_7'] - df['ma_14']) / df['ma_14']

    # RSI 14
    delta = grouped['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.groupby(df['coin']).transform(lambda x: x.rolling(14).mean())
    avg_loss = loss.groupby(df['coin']).transform(lambda x: x.rolling(14).mean())
    rs = avg_gain / avg_loss
    df['RSI_14'] = 100 - (100 / (1 + rs))

    # Target: 1 if Next Day Close > Today's Close
    df['target'] = (grouped['Close'].shift(-1) > df['Close']).astype(int)

    # Drop artifacts
    df = df.replace([np.inf, -np.inf], np.nan).dropna()
    return df

df_eng = engineer_features(df)
print("Feature Engineering Complete.")
```

The Multi-Model Training Engine

```python
features = ['return_1', 'return_3', 'ma_distance_7', 'volatility_7', 'volume_change', 'ma_crossover', 'RSI_14']
coins = df_eng['coin'].unique()

models_export = {}
accuracies = {}

for coin in coins:
    # Isolate data for this specific token
    coin_df = df_eng[df_eng['coin'] == coin].sort_values('Date')

    # Skip coins with barely any historical data
    if len(coin_df) < 500:
        continue

    n = len(coin_df)
    train_end = int(n * 0.8)

    # Time-based 80/20 split
    train_df = coin_df.iloc[:train_end]
    test_df = coin_df.iloc[train_end:]

    X_train, y_train = train_df[features], train_df['target']
    X_test, y_test = test_df[features], test_df['target']

    # Standard Scaler (Easy to replicate in JS)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train Token-Specific Model
    model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
    model.fit(X_train_scaled, y_train)

    # Evaluate
    acc = accuracy_score(y_test, model.predict(X_test_scaled))
    accuracies[coin] = acc

    # Save EVERYTHING the frontend needs to make predictions
    models_export[coin] = {
        "features_order": features,
        "scaler_means": scaler.mean_.tolist(),      # Needed to scale live data in JS
        "scaler_scales": scaler.scale_.tolist(),    # Needed to scale live data in JS
        "weights": model.coef_[0].tolist(),         # The trained model weights
        "intercept": float(model.intercept_[0]),    # The trained model intercept
        "test_accuracy": acc,
        "training_samples": n
    }

print(f"Successfully trained {len(models_export)} independent models!")
print(f"Average Global Accuracy: {np.mean(list(accuracies.values())):.4f}")
```

Model Performance Visualization

```python
# Sort accuracies descending
sorted_acc = dict(sorted(accuracies.items(), key=lambda item: item[1], reverse=True))

plt.figure(figsize=(14, 7))
sns.barplot(x=list(sorted_acc.keys()), y=list(sorted_acc.values()), palette='coolwarm')
plt.axhline(y=0.5, color='black', linestyle='--', label='50% Baseline (Random Guessing)')
plt.title('Prediction Accuracy by Cryptocurrency Model')
plt.xlabel('Token')
plt.ylabel('Test Set Accuracy')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

Exporting the JSON Dashboard Brain

```python
# Save the master dictionary to JSON
with open('multi_model_data.json', 'w') as f:
    json.dump(models_export, f, indent=4)

print("Exported multi_model_data.json successfully!")
```
